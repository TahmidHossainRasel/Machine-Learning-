```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
rgn = ['southwest', 'southeast', 'northwest', 'northeast']
```

```python
new_rgn = pd.DataFrame(encoded, columns=['rgn'])
```

```python
new_rgn.head()
```

```python
df7.head()
```

```python
df7.region= new_rgn
```
