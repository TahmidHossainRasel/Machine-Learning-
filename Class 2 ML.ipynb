# Global and Local Variables

```python
x=50 #global variable
def myfunction1():
    print('This is Number 1: ',x)
myfunction1()
print('This is number 2: ',x)
```

```python
#here Global variable is 20 and Local variable is 30
x=20
def myfunction1():
    x=30
    print('This is Line Number 1:',x)
myfunction1()
print('This is Line Number 2:',x)
```

```python
def myfunction1():
    global var1
    var1=60  #here x is local variable
    print('This is line number 1:',var1)
myfunction1() 
print('This is line number 2:',var1)
#so x cant be access from outsite the class
```

```python
def myfunction():
    global var11  #declear as global variable
    var11=44
    print('This line is number 1:',var11)
myfunction()
print('This line is number 2:',var11)
    #so we can access from anywhere
```
